### select, select_one 
- 참고 자료 : [링크](https://desarraigado.tistory.com/14)
- CSS selector로 tag 객체를 찾아 반환
- 각각 find(), find_all()에 대응
- 특정 경로의 태그를 객체로 반환하고 싶을 때, find의 경우 반복적으로 코드를 작성해야 하지만, select는 직접 하위 경로를 지정할 수 있기 때문에 간편하다.


```python
from bs4 import BeautifulSoup

tag = "<p class='youngone' id='junu'> Hello World! </p>"
soup = BeautifulSoup(tag)

# 태그 이름만 특정 
print(soup.select_one('p'))

# 태그 class 특정 
print(soup.select_one('.youngone')) 

# 태그 이름과 class 모두 특정 
print(soup.select_one('p.youngone')) 

# 태그 id 특정 
print(soup.select_one('#junu')) 

# 태그 이름과 id 모두 특정 
print(soup.select_one('p#junu')) 

# 태그 이름과 class, id 모두 특정 
print(soup.select_one('p.youngone#junu'))

#find 
soup.find('div').find('p') 
#select 
soup.select_one('div > p')
```

Q. `div:nth-child(1)` : div가 가진 요소들 중에 1번째 요소(children)를 가지고 온다. 

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

m_url = 'https://movie.naver.com/movie/running/current.nhn#'
page = urlopen(m_url)
soup = BeautifulSoup(page, "html.parser")
# movies = soup.select('#content > div.article > div:nth-child(2) > div.lst_wrap > ul > li')
movies = soup.select('#content > div.article > div > div.lst_wrap > ul > li')
print('영화제목\t ','평점\t','참여\t','애매율\t\t','감독\n')
for movie in movies:
    title = movie.select_one(' dl > dt > a').text
    star = movie.select_one('span.num').text
    people = movie.select_one('em').text
    rate = movie.select_one(' div > span.num')
    king = movie.select_one('dd:nth-child(4) > span > a').text
    try:
        print(title+'\t', star+'\t', people+'\t', rate.text+'\t', king+'\n')
    except Exception as e:
        continue
```

#### Q. urlopen 대신 selenium을 쓰는 이유는 무엇인가?

알아야할 것! 서버에서 데이터의 흐름
1) 사용자가 서버에 정보를 요청  
2) 호스팅 회사(1차, SKT/KT/LG)들이 사용자의 요청을 해당 웹사이트(2차, 구글/네이버 등)에 전달  
3) 웹사이트는 정보 제공자(3차 블로그 등)를 통해서 정보를 수합하여 호스팅 회사에게 전달  
- 웹사이트는 트래픽을 적정 수준으로 조절  

일반적인 정보 처리 단계에서는 사용자가 API(chrome, safari 등)을 통해서 1~3차에게 정상적인 요청을 보내게 된다.   
하지만, urlopen을 사용하면 API를 거치지 않고 바로 요청이 되기 때문에 정상적인 요청으로 여겨지지 않게 되고 서버에서 차단당하게 된다.   
selenium을 이용하면 웹드라이버를 통해서 웹페이지를 직접 띄움으로서 "정상적인 사용자인 척"할 수 있으며, 여러 정보를 크롤링 할 때 코드를 통해서 자동적으로 설정할 수 있기 때문에 편의성을 지닌다.   
cf. selenium은 원래 테스트용으로 만들어졌다고 한다.
